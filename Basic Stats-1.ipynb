```python
#Histograms
import seaborn as sns
import matplotlib.pyplot as plt

skew_values = df[numerical_cols].skew()
for col in numerical_cols:
        plt.figure(figsize=(10, 5))
        sns.histplot(dt[col], kde=True, bins=30, color='blue', alpha=0.7)
        plt.title(f'Distribution of {col}')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.show()

        # Analyze distribution
        skewness = dt[col].skew()
        print(f"\nAnalysis for '{col}':")
        print(f"Skewness: {skewness:.2f}")
        
        if skewness > 1:
            print("The distribution is highly positively skewed.")
        elif skewness < -1:
            print("The distribution is highly negatively skewed.")
        elif 0.5 < skewness <= 1 or -1 <= skewness < -0.5:
            print("The distribution is moderately skewed.")
        else:
            print("The distribution is approximately symmetric.")
        
        # Check for outliers using the IQR method
        Q1 = dt[col].quantile(0.25)
        Q3 = dt[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        outliers = dt[(dt[col] < lower_bound) | (dt[col] > upper_bound)][col]
        print(f"Number of outliers: {len(outliers)}")
        if len(outliers) > 0:
            print(f"Outliers:{outliers.values}\n")
        else:
            print("No significant outliers detected.\n")

```

```python
#Standardization of Numerical variables
#Concept of Standardization: Standardization is a data preparation procedure that scales numerical data to have a standard deviation of one and a mean of zero. The dataset is more suited for machine learning techniques that are sensitive to feature magnitudes, such logistic regression, support vector machines, and k-means clustering, thanks to this transformation, which helps to bring all features into the same scale.
standardized_data = dt.copy()
for col in numerical_cols:
    mean = dt[col].mean()
    std = dt[col].std()
    standardized_data[col] = (dt[col] - mean) / std
print("Before Standardization:\n", dt[numerical_cols].head())
print("\n")
print("After Standardization:\n", standardized_data[numerical_cols].head())
```

```python
#Although datasets frequently contain categorical variables , ML models usually require numerical input. These categorical variables are transformed into a numerical representation that machine learning algorithms can comprehend through a preprocessing phase called one-hot encoding.
#One hot encoding
print("\nApplying one-hot encoding to categorical columns...")
encoded_data = pd.get_dummies(dt, columns=categorical_cols, drop_first=True)
encoded_data=encoded_data.astype(int)
print("One-hot encoding applied successfully.")

#Transformed dataset
print("\nTransformed dataset (first 5 rows):")
print(encoded_data.head())
```
