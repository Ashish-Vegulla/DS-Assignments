```python
# Import required libraries
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
#Loading the dataset
df = pd.read_csv("C:\\Users\\Ashish\\Downloads\\Data Science Assignments\\Recommendation System\\anime.csv")
df
```

```python
# Display basic info about the dataset
print("Dataset Head:")
print(df.head())
print("\n")
print("\nDataset Info:")
df.info()
```

```python
# Handle missing values
df.fillna("Unknown", inplace=True)

# Feature Extraction
# Use 'genre' and 'rating' for similarity calculation
df['combined_features'] = df['genre'] + " " + df['rating'].astype(str)
```

```python
# Convert categorical text data into numerical vectors
vectorizer = CountVectorizer()
feature_matrix = vectorizer.fit_transform(df['combined_features'])
```

```python
# Calculate Cosine Similarity
cosine_sim = cosine_similarity(feature_matrix)
```

```python
# Recommendation Function
def recommend_anime(title, top_n=5):
    try:
        # Check if 'title' column exists, otherwise try 'name'
        title_col = 'title' if 'title' in df.columns else 'name'
        idx = df[df[title_col] == title].index[0]
        sim_scores = list(enumerate(cosine_sim[idx]))
        sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
        sim_scores = sim_scores[1:top_n+1]  # Exclude the target anime itself
        anime_indices = [i[0] for i in sim_scores]
        return df.iloc[anime_indices][['title' if 'title' in df.columns else 'name', 'genre', 'rating']] #also changes the output column in case title is not found
    except IndexError:
        return "Anime not found in the dataset."
```

```python
print("Recommendations for 'Monster':")
print(recommend_anime("Monster"))
```

```python
# Evaluation
train_data, test_data = train_test_split(df, test_size=0.2, random_state=42)
```

```python
#Evaluation
def evaluate_recommendation():
    y_true, y_pred = [], []
    # Use 'name' column if 'title' doesn't exist
    title_col = 'title' if 'title' in test_data.columns else 'name'
    for title in test_data[title_col]:
        recommendations = recommend_anime(title, top_n=3)
        if isinstance(recommendations, pd.DataFrame):
            y_true.append(1)
            y_pred.append(1)
        else:
            y_true.append(1)
            y_pred.append(0)

    precision = precision_score(y_true, y_pred, zero_division=1)
    recall = recall_score(y_true, y_pred, zero_division=1)
    f1 = f1_score(y_true, y_pred, zero_division=1)
    return precision, recall, f1

precision, recall, f1 = evaluate_recommendation()
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

```python
#1)Can you explain the difference between user-based and item-based collaborative filtering?

Answer:

1. User-based collaborative filtering:

Focuses on identifying comparable users by their activity or preferences (e.g., ratings, purchases).
Suggests products that people who are similar to you have loved.
For instance, User B will be suggested Item X if User A and User B share similar tastes and User A enjoyed it.
Challenges include sparse data and performance problems with big user populations.

2.Item-based collaborative filtering:

Focuses on using user interactions to identify related things.
Suggests products that are comparable to those the consumer has already found enjoyable or highly rated.
For instance, a user will be suggested Item Y if they like Item X and they are similar.
Greater scalability than user-based filtering and effective in situations of higher user charn.
```

```python
#2)What is collaborative filtering, and how does it work?

Answer:

Definition: Collaborative filtering is a suggestion method that predicts what a certain user might like based on their interactions, preferences, or behavior.

How It Operates:

Data collection: Compile information on user-item interactions, such as ratings, clicks, and sales.
Discovering Commonalities:Calculate similarities between objects (item-based) or users (user-based) using metrics such as Pearson correlation and cosine similarity.

Suggestion:
For user-based: Suggest products that users who are similar to you like.
For item-based: Suggest products that are comparable to those the user has already interacted with.

Principal Benefits:
Does not depend on elements of explicit material.
Can reveal links and patterns in user behavior that are concealed.

Principal Difficulties:
Cold start issue: Having trouble suggesting new products or users.
Data sparsity: Accuracy may be decreased by sparse user-item matrices. 
```
